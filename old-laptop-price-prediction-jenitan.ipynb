## Feature Engineering

1 Feature engineering is the process of creating new features or modifying existing ones in a dataset to make it easier for machine learning models to understand and extract patterns from the data.

2 It's like preparing the ingredients before cooking - you're transforming or creating variables in a way that helps the model to learn and make better predictions.

Here we will create percentage depreciation of the laptop and create a feature indicating whether the laptop has a high or low selling price compared to its actual price.

```python
# Calculate depreciation percentag
df['Depreciation (%)'] =((df['Actual Price (Rs)']- df['Old Laptop Selling Price (Rs)'])/df['Actual Price (Rs)']) * 100

# Create a feature indicating if the selling price is high or low compared to the actual price
df['Price Status'] = df['Old Laptop Selling Price (Rs)'] > df['Actual Price (Rs)']
df['Price Status'] = df['Price Status'].map({True: 'High',False: 'Low'})

print("\nDataset after Feature Engineering:")
display(df)
```

    Exaplnation:
    
1 Depreciation Percentage Calculation: It calculates the depreciation percentage of old laptops. The depreciation percentage is a measure of how much the selling price has decreased compared to the original (actual) price. The result is stored in a new column called 'Depreciation (%)'.

2 Price Status Calculation: It determines whether the selling price of each old laptop is considered high or low compared to the actual price. The result is stored in a new column called 'Price Status'. If the selling price is higher than the actual price, the 'Price Status' is set to 'High'; otherwise, it is set to 'Low'.

3 Printing the Result: The code then prints the DataFrame after these calculations, displaying the original data along with the newly added 'Depreciation (%)' and 'Price Status' columns.

# Explanation:

# Import Libraries:

The code starts by importing necessary libraries - pandas for data manipulation, and LinearRegression and LabelEncoder from scikit-learn for machine learning tasks.

# Label Encoding:

The categorical variables in the dataset (Brand, Processor, Physical Condition, Operating System) are converted into numerical values using label encoding. This is done to make it suitable for machine learning algorithms.

# Define Features and Target:

The features (input variables) for the machine learning model are specified as ['RAM (GB)', 'Storage (GB)', 'Year of Release', 'Age (Years)', 'Screen Damage', 'Charger Availability', 'Keyword Working', 'Battery Health (%)'].
The target variable (output to be predicted) is set as 'Old Laptop Selling Price (Rs)'.

# Extract Features and Target:

The features (X) and the target variable (y) are extracted from the DataFrame (df) using the specified column names.

A linear regression model is instantiated using LinearRegression() from scikit-learn.

# Train Model:

The model is trained using the entire dataset (X and y) using thefit` method.

The code takes user input for various laptop attributes such as RAM, Storage, Year of Release, Age, Screen Damage, Charger Availability, Keyword Working, and Battery Health percentage.

The user input is then organized into a DataFrame (user_input_df) for making predictions.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Assuming df is your DataFrame

# Label Encoding for categorical variables
label_encoder = LabelEncoder()
df['Brand'] = label_encoder.fit_transform(df['Brand'])
df['Processor'] = label_encoder.fit_transform(df['Processor'])
df['Physical Condition'] = label_encoder.fit_transform(df['Physical Condition'])
df['Operating System'] = label_encoder.fit_transform(df['Operating System'])

# Convert 'Storage (GB)' column to numeric by extracting the numerical part
df['Storage (GB)'] = df['Storage (GB)'].astype(float)

# Define features and target variable
features = ['RAM (GB)', 'Storage (GB)', 'Year of Release', 'Age (Years)', 'Screen Damage',
            'Charger Availability', 'Keyword Working', 'Battery Health (%)']

target = 'Old Laptop Selling Price (Rs)'

x = df[features]
y = df[target]

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Train the linear regression model on the training set
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')

# Take user input
user_input = {
    'RAM (GB)': 8,
    'Storage (GB)': 256,
    'Year of Release': 2020,
    'Age (Years)': 2,
    'Screen Damage': 0,  # 0 for No
    'Charger Availability': 1,  # 1 for Yes
    'Keyword Working': 1,  # 1 for Yes
    'Battery Health (%)': 90
}

# Predict the selling price for the user input
user_input_df = pd.DataFrame([user_input])
user_pred = model.predict(user_input_df[features])
print(f'Predicted Selling Price: {user_pred[0]}')
```

# Modify Model Hyperparameters:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# Assuming df is your DataFrame
# Assuming label_encoder, features, and target are already defined

# Label Encoding for categorical variables
label_encoder = LabelEncoder()
df['Brand'] = label_encoder.fit_transform(df['Brand'])
df['Processor'] = label_encoder.fit_transform(df['Processor'])
df['Physical Condition'] = label_encoder.fit_transform(df['Physical Condition'])
df['Operating System'] = label_encoder.fit_transform(df['Operating System'])

# Convert 'Storage (GB)' column to numeric by extracting the numerical part
df['Storage (GB)'] = df['Storage (GB)'].astype(float)

# Define features and target variable
features = ['RAM (GB)', 'Storage (GB)', 'Year of Release', 'Age (Years)', 'Screen Damage',
            'Charger Availability', 'Keyword Working', 'Battery Health (%)']

target = 'Old Laptop Selling Price (Rs)'

x = df[features]
y = df[target]

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Experiment with different normalization methods
scalers = [('StandardScaler', StandardScaler()), ('MinMaxScaler', MinMaxScaler())]

for scaler_name, scaler in scalers:
    # Scale the features using the current scaler
    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled = scaler.transform(x_test)

    # Train the Linear Regression model with scaled features
    model = LinearRegression()
    model.fit(x_train_scaled, y_train)

    # Make predictions on the testing set
    y_pred_scaled = model.predict(x_test_scaled)

    # Evaluate the model with scaled features
    mse_scaled = mean_squared_error(y_test, y_pred_scaled)
    r2_scaled = r2_score(y_test, y_pred_scaled)

    print(f'Method: {scaler_name}')
    print(f'Mean Squared Error: {mse_scaled}')
    print(f'R^2 Score: {r2_scaled}')
    print('-' * 40)
```
